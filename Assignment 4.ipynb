```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
# Load facebook ego dataset
facebook_edges_file = 'datasets/1684.edges'
facebook_ego_graph = nx.read_edgelist(facebook_edges_file)

# Load twitter ego dataset
twitter_edges_file = 'datasets/2029971.edges'
twitter_ego_graph = nx.read_edgelist(twitter_edges_file, create_using=nx.DiGraph())
```

```python
# Betweenness Centrality
betweenness_centrality_fb = nx.betweenness_centrality(facebook_ego_graph)

# Closeness Centrality
closeness_centrality_fb = nx.closeness_centrality(facebook_ego_graph)

# PageRank
pagerank_fb = nx.pagerank(facebook_ego_graph)

# Display the results for each centrality measure
print("Betweenness Centrality for facebook:", betweenness_centrality_fb)
print("Closeness Centrality for facebook:", closeness_centrality_fb)
print("PageRank for facebook:", pagerank_fb)
```

```python
# Betweenness Centrality
betweenness_centrality_t = nx.betweenness_centrality(twitter_ego_graph)

# Closeness Centrality
closeness_centrality_t = nx.closeness_centrality(twitter_ego_graph)

# PageRank
pagerank_t = nx.pagerank(twitter_ego_graph)

# Display the results for each centrality measure
print("Betweenness Centrality for twitter:", betweenness_centrality_t)
print("Closeness Centrality for twitter:", closeness_centrality_t)
print("PageRank for twitter:", pagerank_t)
```

(b) Please compare the results in (a). What cause the difference between the results?


1. Betweenness Centrality: Nodes with high betweenness often act as bridges between different clusters or communities. These nodes may not be very close to others or have many incoming connections, but they are critical for information flow.

2. Closeness Centrality: This favors nodes that are centrally located in the graph, with short average distances to all other nodes. Nodes that are close to many others will score high here, even if they aren’t critical for connecting different parts of the network.

3. PageRank: This is different from both betweenness and closeness. It accounts for influence based on the number and quality of incoming edges. A node with fewer but highly connected neighbors can score highly, unlike betweenness or closeness, which focus on direct connectivity.

* A node with high betweenness but low PageRank may be a crucial connector in the network but doesn't have a large following or influence.
* A node with high PageRank but low betweenness may be highly influential but not necessarily a connector or centrally located.
* A node with high closeness but low betweenness may be easily reachable from others but doesn’t serve as a key transit point.

**Causes of Differences:**

*Structural Roles:* Nodes serving different structural roles (e.g., bridges between clusters vs. internal hubs) will have different centrality scores. A node with high betweenness might sit between two otherwise loosely connected groups, but may not be directly central in terms of proximity (closeness) or inbound link strength (PageRank).

*Path Dependency:* Betweenness depends on paths passing through a node, while closeness is more about overall proximity to other nodes. A node that is strategically positioned to connect two regions might score highly on betweenness but could have relatively low closeness if it is not central within any specific region.

*Importance of Connections (PageRank):* PageRank emphasizes not just the quantity but also the quality of connections (i.e., connections to highly ranked nodes). A node could have a high PageRank if it is linked to other important nodes, even if it is not a bridge or particularly close to many nodes.

(d) Choose the results of one of the two algorithms mentioned in question (a), and choose the node
with the highest value and the node with lowest value. Please plot the distributions of Jaccard
similarity, cosine similarity and normalized Euclidean distance between either of the two nodes
and the rest of the nodes in the graph.

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import euclidean

# We are taking Pagerank for twitter-ego dataset
# Get the node with the highest and lowest PageRank values
node_max = max(pagerank_t, key=pagerank_t.get)
node_min = min(pagerank_t, key=pagerank_t.get)

# Function to compute Jaccard similarity between two nodes
def jaccard_similarity(node1, node2):
    neighbors1 = set(twitter_ego_graph.neighbors(node1))
    neighbors2 = set(twitter_ego_graph.neighbors(node2))
    intersection = len(neighbors1 & neighbors2)
    union = len(neighbors1 | neighbors2)
    return intersection / union if union != 0 else 0

# Function to compute Normalized euclidean distance between two nodes
def normalized_euclidean_distance(node1, node2):
    neighbors1 = np.array([1 if i in twitter_ego_graph.neighbors(node1) else 0 for i in twitter_ego_graph.nodes()])
    neighbors2 = np.array([1 if i in twitter_ego_graph.neighbors(node2) else 0 for i in twitter_ego_graph.nodes()])
    return euclidean(neighbors1, neighbors2) / len(twitter_ego_graph.nodes())

# Calculate similarities/distance for both nodes (min and max)
def compute_metrics(node):
    jaccard_similarities = []
    cosine_similarities = []
    euclidean_distances =[]

    node_vec = np.array([1 if i in twitter_ego_graph.neighbors(node) else 0 for i in twitter_ego_graph.nodes()])
    for other_node in twitter_ego_graph.nodes():
        if other_node != node:
            #Jaccard similarity
            jaccard_sim = jaccard_similarity(node, other_node)
            jaccard_similarities.append(jaccard_sim)

            #Cosine similarity
            other_node_vec = np.array([1 if i in twitter_ego_graph.neighbors(other_node) else 0 for i in twitter_ego_graph.nodes()])
            cos_sim = cosine_similarity([node_vec], [other_node_vec])[0][0]
            cosine_similarities.append(cos_sim)

            # Normalized Euclidean distance
            euclidean_dist = normalized_euclidean_distance(node, other_node)
            euclidean_distances.append(euclidean_dist)

    return jaccard_similarities, cosine_similarities, euclidean_distances

# Compute metrics for node_max and node_min
jaccard_max, cosine_max, euclidean_max = compute_metrics(node_max)
jaccard_min, cosine_min, euclidean_min = compute_metrics(node_min)
```

```python
#Plot distributions for both nodes

def plot_distributions(jaccard_vals, cosine_vals, euclidean_vals, title):
    plt.figure(figsize=(18, 5))

    # Jaccard similarity plot
    plt.subplot(1, 3, 1)
    plt.hist(jaccard_vals, bins=30, color='skyblue')
    plt.title(f'Jaccard Similarity Distribution - {title}')
    plt.xlabel('Jaccard Similarity')
    plt.ylabel('Frequency')

    # Cosine similarity plot
    plt.subplot(1, 3, 2)
    plt.hist(cosine_vals, bins=30, color='lightgreen')
    plt.title(f'Cosine Similarity Distribution - {title}')
    plt.xlabel('Cosine Similarity')
    plt.ylabel('Frequency')

    # Normalized Euclidean distance plot
    plt.subplot(1, 3, 3)
    plt.hist(euclidean_vals, bins=30, color='salmon')
    plt.title(f'Normalized Euclidean Distance Distribution - {title}')
    plt.xlabel('Normalized Euclidean Distance')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Plot for the node with highest value (node_max)
plot_distributions(jaccard_max, cosine_max, euclidean_max, 'Node with Highest PageRank')

# Plot for the node with lowest value (node_min)
plot_distributions(jaccard_min, cosine_min, euclidean_min, 'Node with Lowest PageRank')
```
